# Meaning of each column

**The datasheet represents 'most runs' done by any cricketer in One-day internationals**
1. In the first column we get to see the players name
2. Second column is the year span of the each cricketers
3. Third column indicates the matches they have played in their span
4. Fourth column is the number of the innings they have played
5. Fifth column is the number of Not-Outs occured in their span
6. Sixth column is the total runs they have scored
7. Seventh column is the highest score of individual players
8. The batting average is showed in the eighth column
9. Ninth column is the number of 'Ball faced' by a player
10. Tenth column is the Strike Rate
11. Eleventh column is the total number of 100's scored by a player
12. The next column is the total number of 50's scored by a player
13. Thirteenth column represents the number of 'No run' occured by a player
14. Last two columns shows the numbers of 4's and 6's hitted by a player respectively.

# Renaming the column names appropriately

```python
dfbatsman.rename(columns={'Mat':'Match', 
                   'Inns':'Innings',
                   'NO': 'NotOut',
                   'HS': 'Highest_score',
                   'Ave': 'Average',
                   'BF': 'Balls_Faced',
                   'SR': 'Strike_Rate',
                   100: 'Centuries',
                   50: 'Half_centuries',
                   0: 'Ducks',
                   "4s": "Fours",
                   "6s": "Sixes"}, 
          inplace = True)
```

```python
display(dfbatsman.head())
```

# How many players played for ICC?

Doing some data preprocessing before answering this question.

```python
# splitting Player into player's name and country
dfbatsman[["Player_Name", "Country"]] = dfbatsman['Player'].str.split("(", expand=True)  
# removing Player column
dfbatsman = dfbatsman.drop('Player', axis=1)
# removing trailing ) from country
dfbatsman['Country'] = dfbatsman['Country'].str.replace(")", "")
# changing column order
new_col_sequence = ['Player_Name', 'Country', 'Span', 'Match', 'Innings', 'NotOut', 'Runs', 'Highest_score',
       'Average', 'Balls_Faced', 'Strike_Rate', 'Centuries', 'Half_centuries', 'Ducks', 'Fours', 'Sixes']
dfbatsman = dfbatsman[new_col_sequence]
# displaying the new dataset
display(dfbatsman.head())
```

**Number of players played for ICC:**

```python
# defining a function
def icc_check(x):
    res = "Played for ICC" if ("ICC" in x) else "Didn't play for ICC"
    return res

# counting the players
dfbatsman['played_for_ICC'] = dfbatsman['Country'].apply(icc_check)
print(dfbatsman['played_for_ICC'].value_counts())
```

# How many different countries are present in this dataset? 

```python
# observing the unique values 
dfbatsman['Country'].value_counts()
```

```python
# removing chunk of letters that do not mean any countries
dfbatsman['Country'] = dfbatsman['Country'].str.replace("ICC", "")
dfbatsman['Country'] = dfbatsman['Country'].str.replace("Asia", "")
dfbatsman['Country'] = dfbatsman['Country'].str.replace("Afr", "")
dfbatsman['Country'] = dfbatsman['Country'].str.replace("/", "")
dfbatsman['Country'].value_counts()
```

```python
country_num = dfbatsman['Country'].nunique()

print(f"Number of countries present in the dataset = {country_num}")
```

# Which player(s) had played for the longest period of time?

```python
# splitting the 'Span' column based on the "-"
dfbatsman[['start_year', 'end_year']] = dfbatsman['Span'].str.split("-", expand=True).astype('int') 

# removing the "Span" column
dfbatsman = dfbatsman.drop("Span", axis=1)

display(dfbatsman.head())
dfbatsman.dtypes
```

```python
dfbatsman["years_played"] = dfbatsman["end_year"] - dfbatsman["start_year"]
dfbatsman.sort_values("years_played", ascending=False)[['Player_Name','years_played']].head()
```

So the player is **SR Tendulkar**.

```python
dfbatsman.loc[dfbatsman['years_played'].idxmax()]
```

# Which player(s) had played for the shortest period of time?

```python
dfbatsman.sort_values("years_played", ascending=False)[['Player_Name','years_played']].tail()
```

```python
dfbatsman.loc[dfbatsman['years_played'].idxmin()]
```

So the player is **AJ Finch**.

# How many Australian Bowlers are present in this dataset?

```python
dfbowlers = pd.read_excel('ODI_cricket.xlsx', sheet_name="bowler")
display(dfbowlers.head(10))
```

```python
dfbowlers.rename(columns={'Mat':'Match', 
                   'Inns':'Innings',
                   'Wkts': 'Wickets',
                   'Ave': 'Average',
                   'Econ': 'Economy_Rate',
                   'SR': 'Strike_Rate',
                    4 : "Four_wickets",
                    5 : "Five_wickets"}, 
          inplace = True)
display(dfbowlers.head(10))
```

```python
# splitting Player into player's name and country
dfbowlers[["Player_Name", "Country"]] = dfbowlers['Player'].str.split("(", expand=True)  
# removing Player column
dfbowlers = dfbowlers.drop('Player', axis=1)
# removing trailing ) from country
dfbowlers['Country'] = dfbowlers['Country'].str.replace(")", "")
# changing column order
new_col_sequence = ['Player_Name', 'Country', 'Span', 'Match', 'Innings', 'Balls', 'Runs', 'Wickets',
       'Average', 'Economy_Rate', 'Strike_Rate', 'Four_wickets', 'Five_wickets']
dfbowlers = dfbowlers[new_col_sequence]
# displaying the new dataset
display(dfbowlers.head())
```

```python
# observing the unique values 
dfbowlers['Country'].value_counts()
```

```python
# removing chunk of letters that do not mean any countries
dfbowlers['Country'] = dfbowlers['Country'].str.replace("ICC", "")
dfbowlers['Country'] = dfbowlers['Country'].str.replace("Asia", "")
dfbowlers['Country'] = dfbowlers['Country'].str.replace("Afr", "")
dfbowlers['Country'] = dfbowlers['Country'].str.replace("/", "")
dfbowlers['Country'].value_counts()
```
